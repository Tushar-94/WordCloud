# Word Cloud for Atheism

```python
import lime
import sklearn 
import numpy as np
import sklearn 
import sklearn.ensemble
import matplotlib.pyplot as plt 
import sklearn.metrics
import pandas as pd
import os
from wordcloud import WordCloud, STOPWORDS 
from PIL import Image
from wordcloud import ImageColorGenerator

data_tr = fetch_20newsgroups(subset = 'train', categories = ['alt.atheism'])

vectorizer = sklearn.feature_extraction.text.TfidfVectorizer(lowercase = False)
train_vectors = vectorizer.fit_transform(data_tr)
```

# Preparing train data for Wordcloud

```python
train_data_df = pd.DataFrame(data = data_tr.data, columns = ['text'])
train_data_df['target'] = data_tr.target
train_data_df[train_data_df['target'] == 0].shape
```

```python
train_data_df[train_data_df['target'] == 1].shape
```

```python
char_mask = np.array(Image.open("krab.png"))  
image_colors = ImageColorGenerator(char_mask)
```

# Word Cloud Train Data(Class 0)

```python
news_words = ''
stopwords = set(STOPWORDS)
for val in train_data_df[train_data_df['target'] == 0].text:
    # typecast each val to string
    val = str(val)
    # split the values
    tokens = val.split()
    # Converts each token to lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    news_words += " ".join(tokens)+ " "
     
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10,
                mask=char_mask,
                max_words=400
                     ).generate(news_words)
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
#plt.imshow(wordcloud.recolor(color_func=image_colors))
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.title('Atheism',fontsize=50)
  
plt.show()
```
